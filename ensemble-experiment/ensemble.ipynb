### 1. Max Voting: 

```python
allcls = pd.concat([mfcc,specgram,w2vb,w2vl,roberta,roberta_lfl], axis=1)
allcls
```

```python
allcls.iloc[0]
```

```python
#{'negative': 0, 'neutral': 1, 'positive': 2}
# sum of all predcitions
allcls['0'] = (allcls[['predicted_mfcc','predicted_specgram','predicted_w2vb', 'predicted_w2vl', 
                           'predicted_roberta', 'predicted_roberta_lfl']] == 0).sum(axis=1)
allcls['1'] = (allcls[['predicted_mfcc','predicted_specgram','predicted_w2vb', 'predicted_w2vl', 
                           'predicted_roberta', 'predicted_roberta_lfl']] == 1).sum(axis=1)
allcls['2'] = (allcls[['predicted_mfcc','predicted_specgram','predicted_w2vb', 'predicted_w2vl', 
                           'predicted_roberta', 'predicted_roberta_lfl']] == 2).sum(axis=1)
```

```python
#most number of occurrences across all classifiers
allcls['ensemble_maxvoted'] = (allcls[['0', '1', '2']]).idxmax(axis=1)
```

```python
#Identifying equal cases and naming them as "equal case"
for index, row in allcls.iterrows():
    if allcls['0'][index] == 2 and allcls['1'][index] == 2 and allcls['2'][index] == 2:
        allcls['ensemble_maxvoted'][index] = "1" 
    
    elif allcls['0'][index] == 3 and allcls['1'][index] == 3:
            allcls['ensemble_maxvoted'][index] = "1"
        
    elif allcls['0'][index] == 3 and allcls['2'][index] == 3:
            allcls['ensemble_maxvoted'][index] = "2"
        
    elif allcls['1'][index] == 3 and allcls['2'][index] == 3:
            allcls['ensemble_maxvoted'][index] = "2"
```

```python
allcls
```

```python
#total number of equal cases
allcls.loc[allcls['ensemble_maxvoted'] == 'equal case']
```

### Evaluating:

```python
voting = allcls['ensemble_maxvoted'].astype(int)
voting = pd.DataFrame(data=voting)
```

```python
for cls in (mfcc,specgram,w2vb,w2vl,roberta,roberta_lfl, voting):
    print(cls.columns[0], ":", accuracy_score(cls, target))
```

### 2. Averaging: 

```python
dfs = [MFCC, SPECGRAM, W2VB, W2VL, ROBERTA, ROBERTA_LFL]
all_classifiers = pd.concat(dfs,axis=1)
```

```python
all_classifiers
```

```python
#'negative': 0, 'neutral': 1, 'positive': 2}
all_classifiers['0'] = (all_classifiers[['Negative_mfcc', 'Negative_specgram', 'Negative_w2vb', 'Negative_w2vl',
                                                'Negative_roberta', 'Negative_roberta_lfl']].sum(axis=1)/6)

all_classifiers['1'] = (all_classifiers[['Neutral_mfcc', 'Neutral_specgram', 'Neutral_w2vb', 'Neutral_w2vl',
                                               'Neutral_roberta', 'Neutral_roberta_lfl']].sum(axis=1)/6)

all_classifiers['2'] = (all_classifiers[['Positive_mfcc', 'Positive_specgram', 'Positive_w2vb','Positive_w2vl',
                                                'Positive_roberta', 'Positive_roberta_lfl']].sum(axis=1)/6)
```

```python
all_classifiers['ensemble_average'] = (all_classifiers[['0', '1', '2']]).idxmax(axis=1)
```

```python
all_classifiers
```

### Evaluating:

```python
average = all_classifiers['ensemble_average'].astype(int)
average = pd.DataFrame(data=average)
```

```python
for cls in (mfcc,specgram,w2vb,w2vl,roberta,roberta_lfl,average):
    print(cls.columns[0],":", accuracy_score(cls, target))
```
